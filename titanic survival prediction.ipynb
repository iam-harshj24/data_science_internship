```python
df["Age"]= df["Age"].fillna(df["Age"].mean())
df
```

```python
df.isnull().sum()
```

```python
one_hot_encoder = pd.get_dummies(df.Sex)
one_hot_encoder
```

```python
new_data = pd.concat([df,one_hot_encoder],axis ="columns"
                    )
new_data
```

```python
final_data = new_data.drop(["Sex","female"], axis="columns")
final_data
```

```python
final_data['Fare'] = final_data["Fare"].fillna(final_data['Fare'].mean())
final_data
```
